# Walmart Data-Retail Analysis

```python
# Call in pacakages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read in the dataset from the folder
df = pd.read_csv("Walmart_Store_sales.csv")
```

```python
# Convert the string to datetime
df["Date"] = pd.to_datetime(df["Date"])

# Year
df["Year"] = df["Date"].dt.year

# Month
df["Month"] = df["Date"].dt.month

# Day
df["Day"] = df["Date"].dt.day

# Day of Week, Mon=0, Tue=1, ..., Sun=6
df["DOW"] = df["Date"].dt.dayofweek
df.head()

# Convert Fahrenheit to Celsius
df["Temp(C)"] = (df["Temperature"] - 32) * 5/9
```

##### Find Top 5 and Bottom 5 Stores for Sale

```python
# Top 5
round(df.groupby(["Store"]).agg({"Weekly_Sales":"sum"}),0).sort_values("Weekly_Sales",ascending=False).head(5)
```

```python
# Bottom 5
round(df.groupby(["Store"]).agg({"Weekly_Sales":"sum"}),0).sort_values("Weekly_Sales",ascending=True).tail(5)
```

```python
sales = round(df.groupby(["Store"]).agg({"Weekly_Sales":"sum"}),0).sort_values("Weekly_Sales",ascending=False)
```

```python
# Find the max and min sales
max_sales = sales.loc[20]
min_sales = sales.loc[33]
display(max_sales, min_sales)
```

```python
difference_between_highest_and_lowest = max_sales-min_sales
display(difference_between_highest_and_lowest)
```

```python
# Which store has maximum standard deviation i.e., the sales vary a lot. Also, find out the coefficient of mean to standard deviation
mean_std = round(df.groupby(["Store"]).agg(Weekly_mean=('Weekly_Sales','mean'), Weekly_std=("Weekly_Sales","std")),0).sort_values(by=["Weekly_mean","Weekly_std"], ascending=False)
mean_std["Percentage"] = (mean_std["Weekly_std"]/mean_std["Weekly_mean"])*100
mean_std.sort_values("Percentage",ascending=False).head(10)
```

```python
# Which store/s has good quarterly growth rate in Q3’2012
df.loc[df["Month"].isin([1,2,3]), "Quarter"] = "Q1"
df.loc[df["Month"].isin([4,5,6,]), "Quarter"] = "Q2"
df.loc[df["Month"].isin([7,8,9]), "Quarter"] = "Q3"
df.loc[df["Month"].isin([10,11,12]), "Quarter"] = "Q4"
```

```python
df
```

```python
Quarter = df.groupby(["Quarter","Store","Year"]).agg({"Weekly_Sales":"sum"}).reset_index()
Quarter
```

```python
# Q3 2012 and 2012 dataset
Q3_2011 = Quarter[(Quarter["Year"] == 2011) & (Quarter["Quarter"].isin(["Q1"]))] # ,"Q2","Q3"
Q3_2012 = Quarter[(Quarter["Year"] == 2012) & (Quarter["Quarter"].isin(["Q1"]))] # ,"Q2","Q3"
```

```python
# Q3 Weekly Sales 2011 vs 2012
Q3_WS11 = Q3_2011["Weekly_Sales"]
Q3_WS12 = Q3_2012["Weekly_Sales"]
```

```python
WS = ((Q3_WS12.reset_index().drop("index",axis=1)-Q3_WS11.reset_index().drop("index",axis=1))/(Q3_WS11.reset_index().drop("index",axis=1)))*100
```

```python
WS[WS["Weekly_Sales"]==max(WS["Weekly_Sales"])]
```

Store 37 has the highest growth rate in Q3' 2012.

```python
# Make Each Store/ Quarter data into a dictionary format
df_dict = {}
for i in range(45):
    df_dict[i] = Quarter[Quarter["Store"] == Quarter["Store"].unique()[i]]
```

```python
# Plot all the individual stores with total week sales in quarterly fashion
f, axes = plt.subplots(15, 3, figsize=(20,70), sharex=False, sharey=True)
j, k= 0, 0
for i in range(45):
    if i % 3 ==0 and i != 0:
        j += 1
        k = 0
    axes[j, k].set_title("Store" + str(i+1))
#     axes[j, k].set_ylabel("")
    sns.barplot(x="Quarter",y="Weekly_Sales",data=df_dict[i], palette="Set3", hue="Year", ax = axes[j,k])
    k +=1

f.tight_layout(pad=5.0)

# Set Y labels
axes = axes.flatten()
for ax in axes:
    ax.set_ylabel('Dollar $')
```

Most of the stores are at the higest sales in Q4.   
With 2010 Q1 and 2012 Q4 barplots, all the months for that quarter were not captured; thus it has lower sum of weekly sales for those months.
